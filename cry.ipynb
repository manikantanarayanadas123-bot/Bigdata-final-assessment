```python
# === UNITY CATALOG SAFE VERSION ===

table_name = "workspace.default.top_500_metadata2"

# Check table existence using try/except (UC-compatible)
try:
    desc = spark.sql(f"DESCRIBE DETAIL {table_name}")
    exists = True
except Exception as e:
    exists = False

print(f"Table exists? {exists}")

if exists:
    display(desc)

    # Extract Delta table location
    row = desc.select("location").first()
    location = row["location"] if row else None

    print("\n=== PHYSICAL DELTA TABLE LOCATION ===")
    print(location)
else:
    print(f"Table not found: {table_name}")
```

```python
schema = "workspace.default"

tables = spark.sql(f"SHOW TABLES IN {schema}")
table_list = [row.tableName for row in tables.collect()]

results = []
for t in table_list:
    full = f"{schema}.{t}"
    try:
        d = spark.sql(f"DESCRIBE DETAIL {full}")
        loc = d.select("location").first()["location"]
    except:
        loc = None
    results.append((t, loc))

df_paths = spark.createDataFrame(results, ["table_name", "delta_path"])
display(df_paths)
```

```python
# === DATA CLEANING & PREPARATION (CORRECTED COLUMN NAMES) ===

table_name = "workspace.default.top_500_metadata2"

# Load table
df = spark.table(table_name)

print("=== Original Schema ===")
df.printSchema()

# 1. Drop duplicate rows
df_clean = df.dropDuplicates()

# 2. Drop rows with critical nulls
df_clean = df_clean.dropna(subset=["symbol", "current_price"])

# 3. Cast numeric columns to correct types
from pyspark.sql.functions import col
df_clean = df_clean.withColumn("market_cap", col("market_cap").cast("double")) \
                   .withColumn("current_price", col("current_price").cast("double")) \
                   .withColumn("total_volume", col("total_volume").cast("double"))

# 4. Optional: create new derived column (market cap rank)
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, desc

window = Window.orderBy(desc("market_cap"))
df_clean = df_clean.withColumn("market_cap_rank", rank().over(window))

print("=== Cleaned Schema ===")
df_clean.printSchema()

print("\n=== Sample Cleaned Data ===")
display(df_clean.limit(10))

# 5. Save cleaned table back to Unity Catalog as Delta
df_clean.write.format("delta").mode("overwrite").saveAsTable("workspace.default.top_500_metadata_clean")
print("\nCleaned table saved as: workspace.default.top_500_metadata_clean")
```

```python
# === SQL ANALYSIS ON CLEANED TABLE ===

clean_table = "workspace.default.top_500_metadata_clean"

# 1 Query: Top 10 cryptocurrencies by market cap
top_market_cap = spark.sql(f"""
SELECT symbol, name, current_price, market_cap, market_cap_rank
FROM {clean_table}
ORDER BY market_cap DESC
LIMIT 10
""")
print("=== Top 10 Cryptocurrencies by Market Cap ===")
display(top_market_cap)

# 2 Query: Average price and total volume by market cap rank buckets
# Create rank buckets: 1-50, 51-100, etc.
bucketed_stats = spark.sql(f"""
SELECT 
    CASE 
        WHEN market_cap_rank BETWEEN 1 AND 50 THEN '1-50'
        WHEN market_cap_rank BETWEEN 51 AND 100 THEN '51-100'
        WHEN market_cap_rank BETWEEN 101 AND 150 THEN '101-150'
        WHEN market_cap_rank BETWEEN 151 AND 200 THEN '151-200'
        WHEN market_cap_rank BETWEEN 201 AND 250 THEN '201-250'
        WHEN market_cap_rank BETWEEN 251 AND 300 THEN '251-300'
        WHEN market_cap_rank BETWEEN 301 AND 350 THEN '301-350'
        WHEN market_cap_rank BETWEEN 351 AND 400 THEN '351-400'
        WHEN market_cap_rank BETWEEN 401 AND 450 THEN '401-450'
        ELSE '451-500'
    END AS rank_bucket,
    AVG(current_price) AS avg_price,
    SUM(total_volume) AS total_volume
FROM {clean_table}
GROUP BY rank_bucket
ORDER BY rank_bucket
""")
print("=== Average Price & Total Volume by Rank Bucket ===")
display(bucketed_stats)
```
